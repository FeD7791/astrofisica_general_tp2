```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8), dpi=200)

all_head=['ID_LEGAC','ID','RA2','DEC2','zspec', 
        'sigma','e_sigma','HDA','HB','MGB','FE5270',
         'FE5335','D4000','FHB','FOIII5007','n','nser_err','logMvir']

all_df = pd.read_fwf('legac.dat', names = all_head)


y=all_df['n']

mask=(y < 6) & (y > 0)

y1=y[mask]

plt.hist(y1, bins='auto', histtype='step', facecolor='g',
               alpha=0.75)

#plt.xlim(0,2.2)
#plt.ylim(-1, 7)

plt.xlabel("n Sersic")
plt.ylabel("N(n Sersic)")

# Mostrar leyenda, grilla y gráfico
plt.show()
```
